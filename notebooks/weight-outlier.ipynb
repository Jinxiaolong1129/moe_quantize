```python
import seaborn as sns
import torch
from matplotlib import pyplot as plt

plt.rcParams['font.family'] = 'Times New Roman'
```

```python
data = torch.randn(10, 1000, dtype=torch.float16)
# data[1, 10] = 10000
```

```python
plt.boxplot(data.abs(), positions=range(len(data)), showfliers=True)
plt.yscale('log')
plt.show()
```
