```python
import seaborn as sns
import torch
from matplotlib import pyplot as plt

plt.rcParams['font.family'] = 'Times New Roman'
```

```python
name_to_max_weight = torch.load("/Users/lipingzhi/Downloads/name_to_max_weight.pt", map_location='cpu')
name_to_mean_weight = torch.load("/Users/lipingzhi/Downloads/name_to_mean_weight.pt", map_location='cpu')
name_to_max_weight_mistral = torch.load("/Users/lipingzhi/Downloads/name_to_max_weight-mistral.pt", map_location='cpu')
name_to_mean_weight_mistral = torch.load("/Users/lipingzhi/Downloads/name_to_mean_weight-mistral.pt", map_location='cpu')
```

```python
name_to_max_weight.keys()
```

```python
layer_id = 31
plt.figure(figsize=(3.5, 2.5))
abs_weights = [name_to_max_weight[name].values.abs().float() for name in name_to_max_weight.keys() if
               f'model.layers.{layer_id}.block_sparse_moe' in name]
plt.boxplot(abs_weights, positions=range(len(abs_weights)), showfliers=True, showmeans=True, meanline=True, meanprops=dict(color="#EF426F"), flierprops={'markeredgecolor': '#13294B'},)
plt.xlabel("Expert")
plt.ylabel("Weight")
# plt.yscale('log')
ax = plt.gca()
ax.yaxis.set_major_locator(plt.MaxNLocator(4))
ax.xaxis.set_major_locator(plt.MaxNLocator(8))
plt.title(f"Block {layer_id} Max Weight per Channel")
plt.tight_layout()
# plt.show()
plt.savefig(f"block_{layer_id}_max_weight_per_channel.jpg")
```
